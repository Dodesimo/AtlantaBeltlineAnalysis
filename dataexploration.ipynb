# Dataset Overview:

Main dataset: demographic data explorer
Collects data about subregions within the Atlanta Beltline.

Within the dataset there are three subdivisions
- Beltline Equity Indicators
    -  Considers income, unemployment, vehicle availability, housing affordability, race/ethnicity.
- Housing affordability and characteristics:
    -  Considers cost burden, household vehicle availability, internet and computer access, housing tenure, median home value, race/ethnicity
-  Economics and education:
    -  Considers educational attainment, unemployment, poverty, income (economics/eduction), race/ethnicity

Each subdivision mentioned contains a CSV file with data about an entire region/city within the beltline, and also change for that metric.

# Issues with the data for our use case:
- Data provides a broad overview of the county at hand and does not provide information about individuals (but this is applicable to all census data).
- Data does not explicitly define where housing units exist or actual prices of each housing unit (there is map data, but that data cannot be analyzed in a CSV format).
- Does not consider amenities.


# Ideas for what data features to use and how to conduct the simulation.

- Obvious indicators:
    - Housing affordability, vehicle affordability, median home value, housing tenure, educational attainment, and income.

How would these be used in the simulation:

- Same initialization procedures as the original paper, but rather than random values, we use the data provided in the tables.
- Housing units would have their prices set based on the distribution/percentages established (% of houses above a certain prices).
- Individuals' distance/access to amenities would be based on vehicle availability proportions.
- Over time steps in the simulation, house prices and other features would change based on the corresponding dataset for changes.
